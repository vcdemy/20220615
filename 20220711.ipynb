```python
import requests
```

```python
url = " https://www.moedict.tw/uni/"
```

```python
data = requests.get(url+"教室").json()
```

```python
type(data)
```

```python
data
```

```python
data['heteronyms'][0]['definitions'][0]['def']
```

```python
def lookup(word):
  url = " https://www.moedict.tw/uni/"
  data = requests.get(url+word).json()
  return data['heteronyms'][0]['definitions'][0]['def']
```

```python
lookup("教室")
```

```python
import gradio as gr
```

```python
gr.Interface(lookup, "text", "text", title="國語辭典").launch(share=True)
```
