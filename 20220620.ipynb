## 讀取圖形檔

```python
from PIL import Image
```

```python
im = Image.open("download.png")
```

```python
im
```

## 讀取圖片的屬性

```python
im.size
```

```python
im.format
```

```python
im.mode
```

## 圖形檔存檔

```python
im_rgb = im.convert("RGB")
```

```python
im_rgb.save("victor.jpg")
```

## 調換色層

```python
r, g, b = im_rgb.split()
im_rgb = Image.merge("RGB", (b, g, r))
```

```python
im_rgb
```

## 圖形縮放

* resize

```python
im_rgb.size
```

```python
im_rgb.resize?
```

```python
im_rgb.resize((im_rgb.size[0]//3, im_rgb.size[1]//3))
```

```python
im_rgb = im_rgb.resize((im_rgb.size[0]//3, im_rgb.size[1]//3))
```

```python
r, g, b = im_rgb.split()
im_rgb = Image.merge("RGB", (b, g, r))
im_rgb
```

## 圖形旋轉 (Image Rotation)

* transpose
* rotate

```python
im_rgb.transpose?
```

```python
im_rgb.transpose(Image.FLIP_LEFT_RIGHT)
```

```python
im_rgb.transpose(Image.ROTATE_90)
```

```python
im_rgb
```

```python
im_rgb.rotate?
```

```python
im_rgb.rotate(45)
```

```python
im_rgb.rotate(45,expand=True)
```

```python
im_rgb.rotate(45,expand=True,fillcolor="white")
```

## 拼接圖形

```python
im1 = Image.open("妹妹1.jpg")
```

```python
im1.mode
```

```python
im1.size
```

```python
im1 = im1.resize((im1.size[0]//4, im1.size[1]//4))
```

```python
im2 = im2.resize((im2.size[0]//4, im2.size[1]//4))
```

```python
im1
```

```python
im2
```

```python
im1.size
```

```python
im2.size
```

```python
new_im
```
